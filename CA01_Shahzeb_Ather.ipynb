Based on the data quality issues identified above, follwing steps were applied for preprocessing.


```python
#Pre-Processing
#check missing values

df.isna().sum()
```

```python
# Identify numerical and categorical columns + IDNESS

num_cols = df.select_dtypes(include=['int64', 'float64']).columns
cat_cols = df.select_dtypes(include=['object']).columns

# Impute categorical missing values
df[cat_cols] = df[cat_cols].fillna('None')

# Impute numerical missing values using median
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
```

```python
df.isna().sum().sum()
```

Several features contain missing values, with varying levels of severity.
Missing values were identified using column-wise null counts, and features with non-zero missing values were summarized to assess the extent of data quality issues.

Missing numerical values were imputed using the median to minimize the impact of skewness and outliers, while categorical missing values were filled with “None” to preserve feature absence. This approach was applied consistently across all features.

```python
# Outlier handling using IQR capping
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df[col] = df[col].clip(lower_bound, upper_bound)
```

Although z-score–based outlier detection can be effective for normally distributed data, the housing dataset exhibits strong skewness and contains valid extreme values. Therefore, IQR-based detection and capping were used instead to reduce the influence of outliers while preserving all observations.


```python
#Normalization and Scaling
#Split numeric vs Categorical and scale numercial values only

from sklearn.preprocessing import StandardScaler

X = df.drop(columns=['SalePrice'])
y = df['SalePrice']

num_cols = X.select_dtypes(include=['int64', 'float64']).columns

scaler = StandardScaler()
X[num_cols] = scaler.fit_transform(X[num_cols])
```

```python
X[num_cols].describe().loc[['mean', 'std']]
```

```python
#dropping 0 variance features
zero_var_cols = X[num_cols].std() == 0
zero_var_cols = zero_var_cols[zero_var_cols].index

X.drop(columns=zero_var_cols, inplace=True)
```

Features with zero variance were removed as they do not contribute predictive information.


```python
#ENCODING df

df = pd.get_dummies(df, drop_first=True)
```

```python
# Collinearity with Numeric features only
numeric_df = df.select_dtypes(include=['int64', 'float64'])
corr = numeric_df.corr()

# Mask upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

plt.figure(figsize=(10, 8))
sns.heatmap(
    corr,
    mask=mask,
    cmap='coolwarm',
    center=0,
    square=True,
    linewidths=0.5,
    cbar_kws={"shrink": 0.8}
)
plt.title("Collinearity Heatmap (Numeric Features Only)", fontsize=14)
plt.tight_layout()
plt.show()
```

To improve interpretability and computational efficiency, collinearity analysis was performed on a subset of key numerical features known to exhibit redundancy. The heatmap highlights strongly correlated feature pairs used to inform feature selection.


### Collinearity Observations & Feature Selection

Highly correlated feature pairs such as **GrLivArea and TotRmsAbvGrd**, and **GarageCars and GarageArea** were identified through correlation analysis. These pairs exhibit strong linear relationships, indicating redundant information. In such cases, one feature from each pair could be removed to reduce multicollinearity while retaining interpretability and predictive value.
